1) Given two integer numbers return their product only if the product is greater than 1000, else return their sum. write a function for this.

image_2021-11-23_120126.png

```python
num1 = input("Please give me a number: ")
num2 = input("Please give me the second a number: ")
def productNumber(num1, num2):
  n1 = int(num1)
  n2 = int(num2)

  addition = n1 + n2
  multiplication = n1 * n2

  if(multiplication > 1000):
    print(multiplication)
  else:
    print(addition)
productNumber(num1, num2)
```
